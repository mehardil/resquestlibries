```python
import requests

resp = requests.get('https://reqres.in/api/users')
status_code = resp.status_code
status_ok = resp.ok
texts = resp.text
contents = resp.content
jsons = resp.json()

if status_ok == True:
    print("response code" ,status_code)
    print("is response good or bad " ,status_ok)
    print(type(texts))
    print(type(contents))
    print(type(jsons))
    
  
else:
    print("teri _ ka")
```
